### 평가: R² Score & RMSE

**MSE → RMSE 관계:**

$$
\mathrm{MSE} = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2
\qquad\Longrightarrow\qquad
\mathrm{RMSE} = \sqrt{\mathrm{MSE}}
$$

- **R²**: 1에 가까울수록 좋음 (모델이 분산을 얼마나 설명하는지)
- **RMSE**: 0에 가까울수록 좋음 (예측 오차의 크기, 타겟과 같은 단위)
