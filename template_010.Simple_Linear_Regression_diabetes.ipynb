## RMSE (Root Mean Squared Error, 평균 제곱근 오차) 계산

RMSE는 예측값과 실제값의 차이를 측정하는 지표입니다. 
$$
\mathrm{RMSE}
=
\sqrt{
\frac{1}{n}
\sum_{i=1}^{n}
\left( y_i - \hat{y}_i \right)^2
}
$$
- n: 샘플 수
- $y_i$: 실제값
- $ŷ_i$: 예측값
RMSE는 0에 가까울수록 좋은 모델이며, 실제값과 같은 단위를 가집니다.
