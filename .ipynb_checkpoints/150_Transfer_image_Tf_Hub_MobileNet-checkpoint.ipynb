# 전이학습을 이용한 꽃 이미지 분류

Pre-trained MobileNetV2 모델을 사용하여 꽃 이미지를 분류하는 모델을 만듭니다.

## 1. 필요한 라이브러리 import

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print(f"TensorFlow version: {tf.__version__}")
print(f"Keras version: {keras.__version__}")
```

## 2. 데이터 로드 및 전처리

TensorFlow에서 제공하는 꽃 데이터셋을 다운로드하고 전처리합니다.

```python
# 꽃 데이터셋 다운로드
dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
data_dir = tf.keras.utils.get_file('flower_photos', origin=dataset_url, untar=True)
data_dir = data_dir.replace('\\', '/')  # Windows 경로 처리

print(f"데이터 경로: {data_dir}")
```

```python
# 이미지 크기 및 배치 크기 설정
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

# ImageDataGenerator를 사용하여 데이터 로드
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    validation_split=0.2,  # 20%를 validation으로 사용
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    zoom_range=0.2
)

# 훈련 데이터
train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    shuffle=True
)

# 검증 데이터
validation_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    shuffle=False
)

print(f"\n클래스 수: {train_generator.num_classes}")
print(f"클래스 이름: {list(train_generator.class_indices.keys())}")
print(f"훈련 샘플 수: {train_generator.samples}")
print(f"검증 샘플 수: {validation_generator.samples}")
```

## 3. 데이터 시각화

로드된 이미지 데이터를 확인합니다.

```python
# 배치 하나 가져오기
sample_images, sample_labels = next(train_generator)
class_names = list(train_generator.class_indices.keys())

# 이미지 시각화
plt.figure(figsize=(12, 8))
for i in range(min(12, len(sample_images))):
    plt.subplot(3, 4, i+1)
    # preprocess_input이 적용된 이미지는 역정규화 필요
    img = sample_images[i]
    img = (img + 1) / 2  # [-1, 1] 범위를 [0, 1]로 변환
    img = np.clip(img, 0, 1)
    plt.imshow(img)
    label_idx = np.argmax(sample_labels[i])
    plt.title(class_names[label_idx], fontsize=10)
    plt.axis('off')
plt.tight_layout()
plt.show()
```

## 4. Pre-trained 모델 로드

ImageNet으로 사전 학습된 MobileNetV2를 feature extractor로 사용합니다.

```python
# Base model 생성 (top layer 제거)
base_model = MobileNetV2(
    input_shape=(*IMG_SIZE, 3),
    include_top=False,  # 분류 레이어 제거
    weights='imagenet'  # ImageNet 사전 학습 가중치 사용
)

# Base model의 가중치는 학습하지 않음 (freeze)
base_model.trainable = False

print("Base model 요약:")
base_model.summary()
```

## 5. 모델 구성

Base model 위에 새로운 분류 레이어를 추가합니다.

```python
# 전체 모델 구성
model = keras.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),  # Feature map을 벡터로 변환
    layers.Dropout(0.2),  # 과적합 방지
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(train_generator.num_classes, activation='softmax', name='predictions')
])

# 모델 컴파일
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("전체 모델 구조:")
model.summary()
```

## 6. 모델 학습

Feature extraction 단계에서 모델을 학습시킵니다.

```python
# 학습 설정
EPOCHS = 10

# 콜백 설정
callbacks = [
    keras.callbacks.EarlyStopping(
        monitor='val_accuracy',
        patience=3,
        restore_best_weights=True
    ),
    keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=2,
        min_lr=1e-7
    )
]

# 모델 학습
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=validation_generator,
    callbacks=callbacks,
    verbose=1
)
```

## 7. 학습 과정 시각화

학습 과정에서의 정확도와 손실을 그래프로 확인합니다.

```python
# 학습 히스토리 시각화
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# 정확도
axes[0].plot(history.history['accuracy'], label='Train Accuracy', marker='o')
axes[0].plot(history.history['val_accuracy'], label='Validation Accuracy', marker='s')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].set_title('Model Accuracy')
axes[0].legend()
axes[0].grid(True)

# 손실
axes[1].plot(history.history['loss'], label='Train Loss', marker='o')
axes[1].plot(history.history['val_loss'], label='Validation Loss', marker='s')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].set_title('Model Loss')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

## 8. 모델 평가

검증 데이터로 모델의 성능을 평가합니다.

```python
# 모델 평가
test_loss, test_accuracy = model.evaluate(validation_generator, verbose=0)
print(f"검증 손실: {test_loss:.4f}")
print(f"검증 정확도: {test_accuracy:.4f}")
```

## 9. 예측 결과 시각화

검증 데이터에 대한 예측 결과를 시각화합니다.

```python
# 검증 데이터에서 배치 가져오기
val_images, val_labels = next(validation_generator)

# 예측 수행
predictions = model.predict(val_images, verbose=0)
predicted_classes = np.argmax(predictions, axis=1)
true_classes = np.argmax(val_labels, axis=1)

# 결과 시각화
plt.figure(figsize=(15, 10))
for i in range(min(16, len(val_images))):
    plt.subplot(4, 4, i+1)
    img = val_images[i]
    img = (img + 1) / 2
    img = np.clip(img, 0, 1)
    plt.imshow(img)
    
    true_label = class_names[true_classes[i]]
    pred_label = class_names[predicted_classes[i]]
    confidence = predictions[i][predicted_classes[i]]
    
    color = 'green' if true_classes[i] == predicted_classes[i] else 'red'
    plt.title(f'True: {true_label}\nPred: {pred_label} ({confidence:.2f})', 
              color=color, fontsize=9)
    plt.axis('off')
plt.tight_layout()
plt.show()
```

## 10. 혼동 행렬 (Confusion Matrix)

각 클래스별 분류 성능을 확인합니다.

```python
from sklearn.metrics import confusion_matrix, classification_report

# 전체 검증 데이터에 대한 예측
val_generator = validation_generator
val_generator.reset()
all_preds = []
all_labels = []

for i in range(len(val_generator)):
    images, labels = next(val_generator)
    preds = model.predict(images, verbose=0)
    all_preds.extend(np.argmax(preds, axis=1))
    all_labels.extend(np.argmax(labels, axis=1))
    if i >= len(val_generator) - 1:
        break

# 혼동 행렬 계산
cm = confusion_matrix(all_labels, all_preds)

# 혼동 행렬 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names,
            yticklabels=class_names,
            cbar_kws={'label': 'Count'})
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt.title('Confusion Matrix - Flower Classification', fontsize=14, pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# 분류 리포트 출력
print("\n분류 리포트:")
print(classification_report(all_labels, all_preds, target_names=class_names))
```

## 11. (선택사항) Fine-tuning

Base model의 일부 레이어를 unfreeze하여 fine-tuning을 수행할 수 있습니다.

```python
# Fine-tuning을 위해 base model의 일부 레이어를 학습 가능하게 설정
base_model.trainable = True

# 마지막 몇 개 레이어만 학습 가능하게 설정
fine_tune_at = len(base_model.layers) - 30
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

# 더 낮은 학습률로 재컴파일
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.0001),  # 더 낮은 학습률
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print(f"Fine-tuning 시작 레이어: {fine_tune_at}")
print(f"학습 가능한 레이어 수: {sum([1 for layer in base_model.layers if layer.trainable])}")

# Fine-tuning 학습
fine_tune_epochs = 5
total_epochs = len(history.history['loss']) + fine_tune_epochs

history_fine = model.fit(
    train_generator,
    epochs=total_epochs,
    initial_epoch=len(history.history['loss']),
    validation_data=validation_generator,
    callbacks=callbacks,
    verbose=1
)

# Fine-tuning 후 평가
test_loss_fine, test_accuracy_fine = model.evaluate(validation_generator, verbose=0)
print(f"\nFine-tuning 후 검증 정확도: {test_accuracy_fine:.4f}")
print(f"이전 정확도: {test_accuracy:.4f}")
print(f"개선: {test_accuracy_fine - test_accuracy:.4f}")
```
